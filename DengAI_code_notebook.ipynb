```python
#importing libraries
import pandas as pd
import numpy as np
```

# Step 4: Load and inspect datasets

```python
#loading the datasets
X_train = pd.read_csv("dengue_features_train.csv")
y_train = pd.read_csv("dengue_labels_train.csv")
X_test  = pd.read_csv("dengue_features_test.csv")
submission = pd.read_csv("submission_format.csv")
```

```python
#inspect the data
X_train.head()
```

# Step 5: Identify target (total_cases)

```python
#inspect the labels (targets)
y_train.head()
```

```python
#checking the data size
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
#merging features and labels
train_df = X_train.merge(
    y_train,
    on=["city", "year", "weekofyear"]
)

train_df.head()
```

# Step 6: Handle missing values

```python
#checking for missing values
train_df.isna().sum().sort_values(ascending=False).head(10)

#output:
#-Missing values exist (expected in real climate data)
#-Most missing:
#---ndvi_ne (194)
#---ndvi_nw (52)
#Some station measurements also missing
```

```python
#HandlİNG missing values - Mean imputation per city
train_df_imputed = train_df.copy()

numeric_cols = train_df_imputed.select_dtypes(include=[np.number]).columns

train_df_imputed[numeric_cols] = (
    train_df_imputed
    .groupby("city")[numeric_cols]
    .transform(lambda x: x.fillna(x.mean()))
)
```

```python
#verifying that missing values are gone
train_df_imputed.isna().sum().sort_values(ascending=False).head()
```

# Step 7: Separate data by city (San Juan / Iquitos)

```python
train_sj = train_df_imputed[train_df_imputed["city"] == "sj"].sort_values("week_start_date")
train_iq = train_df_imputed[train_df_imputed["city"] == "iq"].sort_values("week_start_date")

print("San Juan shape:", train_sj.shape)
print("Iquitos shape:", train_iq.shape)
```

# Step 8: Time-based train/validation split

```python
#First 80% → training
#Last 20% → validation

def time_split(df, split_ratio=0.8):
    split_idx = int(len(df) * split_ratio)
    train = df.iloc[:split_idx]
    val   = df.iloc[split_idx:]
    return train, val

sj_train, sj_val = time_split(train_sj)
iq_train, iq_val = time_split(train_iq)

print("SJ train:", sj_train.shape, "SJ val:", sj_val.shape)
print("IQ train:", iq_train.shape, "IQ val:", iq_val.shape)
```

# Step 9: Prepare features & target (Model-ready)

```python
drop_cols = ["city", "week_start_date"]

features = [c for c in sj_train.columns if c not in drop_cols + ["total_cases"]]

X_sj_train = sj_train[features]
y_sj_train = sj_train["total_cases"]

X_sj_val = sj_val[features]
y_sj_val = sj_val["total_cases"]

X_iq_train = iq_train[features]
y_iq_train = iq_train["total_cases"]

X_iq_val = iq_val[features]
y_iq_val = iq_val["total_cases"]

print("Number of features:", len(features))
```

# Step 10: Train Polynomial Regression model


Model 1: Polynomial Regression (San Juan first)

```python
#start with degree = 2
#evaluate with MAE
#later we can try degree 3, 4 and compare

#import some libraries
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
#Train Polynomial Regression (San Juan, degree = 2)
poly = PolynomialFeatures(degree=2, include_bias=False)

X_sj_train_poly = poly.fit_transform(X_sj_train)
X_sj_val_poly   = poly.transform(X_sj_val)

model = LinearRegression()
model.fit(X_sj_train_poly, y_sj_train)

y_sj_val_pred = model.predict(X_sj_val_poly)

mae_sj = mean_absolute_error(y_sj_val, y_sj_val_pred)
mae_sj
```

```python
#Train Polynomial Regression for Iquitos (degree = 2)
poly_iq = PolynomialFeatures(degree=2, include_bias=False)

X_iq_train_poly = poly_iq.fit_transform(X_iq_train)
X_iq_val_poly   = poly_iq.transform(X_iq_val)

model_iq = LinearRegression()
model_iq.fit(X_iq_train_poly, y_iq_train)

y_iq_val_pred = model_iq.predict(X_iq_val_poly)

mae_iq = mean_absolute_error(y_iq_val, y_iq_val_pred)
mae_iq
```

for k = 2

Iquitos is much easier to predict

San Juan has higher variance / complexity

# Step 11: Choose and justify polynomial degree (k)

```python
# here we try for k=2, k=3, k=4 to find the minimum polynomial degree

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

def eval_polynomial_degrees(X_train, y_train, X_val, y_val, degrees):
    results = {}
    for d in degrees:
        poly = PolynomialFeatures(degree=d, include_bias=False)

        X_train_poly = poly.fit_transform(X_train)
        X_val_poly   = poly.transform(X_val)

        model = LinearRegression()
        model.fit(X_train_poly, y_train)

        y_pred = model.predict(X_val_poly)
        mae = mean_absolute_error(y_val, y_pred)

        results[d] = mae
    return results

degrees = [2, 3, 4]

sj_results = eval_polynomial_degrees(
    X_sj_train, y_sj_train,
    X_sj_val, y_sj_val,
    degrees
)

iq_results = eval_polynomial_degrees(
    X_iq_train, y_iq_train,
    X_iq_val, y_iq_val,
    degrees
)

print("San Juan MAE:", sj_results)
print("Iquitos MAE:", iq_results)
```

Results Summary

San Juan (MAE):

* Degree 2 → 53.52  [1st] (best)

* Degree 3 → 303.77 [2nd]

* Degree 4 → 646.93 [3rd]

Iquitos (MAE):

* Degree 2 → 15.41 [1st] (best)

* Degree 3 → 46.05 [2nd]

* Degree 4 → 55.96 [3rd]

In Conclusion:

* Higher-degree polynomials severely overfit

* Error explodes as degree increases

Degree = 2 balances:

* model capacity

* data size

* generalization

# Step 20 : Build time-ahead forecasting models
Step 21 : Predict 1, 2, 3+ weeks ahead [DONE]

Step 22 : Evaluate MAE for each horizon [DONE]

Step 23 : Select best prediction horizon [DONE]


```python
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import GradientBoostingRegressor

def make_future_target(df, horizon):
    df = df.sort_values("week_start_date").copy()
    df["future_cases"] = df["total_cases"].shift(-horizon)
    return df.dropna(subset=["future_cases"])

def time_split(df, ratio=0.8):
    split = int(len(df) * ratio)
    return df.iloc[:split], df.iloc[split:]

def eval_horizons(df, horizons):
    results = {}
    for h in horizons:
        df_h = make_future_target(df, h)
        tr, va = time_split(df_h)

        Xtr, ytr = tr[features], tr["future_cases"]
        Xva, yva = va[features], va["future_cases"]

        model = GradientBoostingRegressor(
            n_estimators=300,
            learning_rate=0.05,
            max_depth=3,
            random_state=42
        )
        model.fit(Xtr, ytr)
        pred = model.predict(Xva)

        results[h] = mean_absolute_error(yva, pred)
    return results

horizons = [1, 2, 3, 4]

sj_horizon_mae = eval_horizons(train_sj, horizons)
iq_horizon_mae = eval_horizons(train_iq, horizons)

print("San Juan MAE by horizon:", sj_horizon_mae)
print("Iquitos MAE by horizon:", iq_horizon_mae)
```

Results

San Juan (MAE):

* 1 week → 17.54 [1st] (best)

* 2 weeks → 17.82 [2nd]

* 3 weeks → 18.25 [3rd]

* 4 weeks → 18.72 [4th]

Iquitos (MAE):

* 1 week → 7.57 [1st] (best)

* 2 weeks → 9.15 [2nd]

* 3 weeks → 8.06 [3rd]

* 4 weeks → 11.17 [4th]

Error increases as horizon grows

1-week-ahead forecasting is most reliable

# STEP 24 — Train best model on FULL training data (weekly prediction)

```python
# Prepare full training sets (weekly prediction)
X_sj_full = train_sj[features]
y_sj_full = train_sj["total_cases"]

X_iq_full = train_iq[features]
y_iq_full = train_iq["total_cases"]

# Train final models
final_gbr_sj = GradientBoostingRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)
final_gbr_iq = GradientBoostingRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

final_gbr_sj.fit(X_sj_full, y_sj_full)
final_gbr_iq.fit(X_iq_full, y_iq_full)
```

```python
# Copy test data
test_df = X_test.copy()

# Identify numeric columns (same as training features)
numeric_cols = features

# Apply per-city mean imputation on TEST data
test_df[numeric_cols] = (
    test_df
    .groupby("city")[numeric_cols]
    .transform(lambda x: x.fillna(x.mean()))
)

# Sanity check
test_df[numeric_cols].isna().sum().sum()
```

#Step 25: Generate test predictions

```python
# Split test data by city
test_sj = test_df[test_df["city"] == "sj"].copy()
test_iq = test_df[test_df["city"] == "iq"].copy()

X_test_sj = test_sj[features]
X_test_iq = test_iq[features]

# Predict
test_sj["total_cases"] = final_gbr_sj.predict(X_test_sj)
test_iq["total_cases"] = final_gbr_iq.predict(X_test_iq)

# Combine into submission format
submission_df = pd.concat([test_sj, test_iq])[
    ["city", "year", "weekofyear", "total_cases"]
].sort_values(["city", "year", "weekofyear"])

submission_df.head()
```
